```python
#@title
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dataCafe.csv')

x = df['Fecha']
y = df['Último']

plt.plot(x, y, color='b', linewidth=2, label="Último")
plt.xlabel('Fecha')
plt.ylabel('Último')
plt.title('Titulo')
plt.legend()
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python

df.index = pd.to_datetime(x)

df.index

df.plot()


```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
ete = seasonal_decompose(df['Último'], model = 'aditive')

ete.trend.plot()
```
